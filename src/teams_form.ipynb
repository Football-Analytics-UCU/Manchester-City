```python
import pandas as pd
from soccerbars import soccerbar
```

```python
df = pd.read_csv('../data/FBRef/matches_season_2023-2024.csv')
df = df.drop(columns=['Link', 'Home Team ID', 'Away Team ID', 'Home xG', 'Away xG', 'Home npxG', 'Away npxG', 'Home xAG', 'Away xAG', 'Shots', 'Home Player Stats', 'Away Player Stats', 'Home Ast', 'Away Ast'])
df = df.loc[df['Date'] < '2023-11-12']
mcity = df.loc[df['Away Team'].isin(['Manchester City']) | df['Home Team'].isin(['Manchester City'])]
mcity
```

```python
chelsea = df.loc[df['Away Team'].isin(['Chelsea']) | df['Home Team'].isin(['Chelsea'])]
chelsea
```

```python
isaway = []
home_goals = []
away_goals = []
for index, row in mcity.iterrows():
    if row['Away Team'] == 'Manchester City':
        isaway.append(True)
    else: isaway.append(False)
    home_goals.append(row['Home Goals'])
    away_goals.append(row['Away Goals'])

soccerbar([
    home_goals, 
    away_goals, 
    isaway
], outlined=True)
```
