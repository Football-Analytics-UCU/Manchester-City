```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch, VerticalPitch
import seaborn as sns
```

```python
df = pd.read_csv('events_1729243.csv')
```

```python
df_shorten = df[['period', 'minute', 'second', 'type', 'outcome_type', 'team_id', 'team', 'x', 'y']]
```

```python
df_shorten
```

```python
df_shots = df_shorten[df_shorten['type'].isin(['MissedShots', 'SavedShot', 'Goal'])]
```

```python
df_shots
```

```python
df_shots['type'] = df_shots['type'].replace({'MissedShots': 'MissedShot'})
```

```python
df_shots
```

```python
df_shots.info()
```

```python
chelsea_shots = df_shots[df_shots['team'] == 'Chelsea']
```

```python
#Convert the data into mplsoccer pitch
chelsea_shots ['x'] = chelsea_shots ['x']*1.2
chelsea_shots ['y'] = chelsea_shots ['y']*.8
```

```python
# Create the pitch
fig, ax = plt.subplots(figsize=(13.5, 8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

# Draw the pitch
pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc', half=True)
pitch.draw(ax=ax)

plt.gca().invert_xaxis()

# Plot shots on the pitch
colors = {'MissedShot': 'red', 'SavedShot': 'blue', 'Goal': 'green'}
for shot_type, color in colors.items():
    subset = chelsea_shots[chelsea_shots['type'] == shot_type]
    ax.scatter(subset['x'], subset['y'], color=color, label=shot_type, zorder=5)
    
ax.legend()
plt.title('Chelsea_Shots', c='w', fontsize=14)
```
